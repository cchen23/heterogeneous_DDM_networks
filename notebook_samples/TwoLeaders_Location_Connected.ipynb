```python
import numpy as np
import DDM_util as util
import matplotlib.pyplot as plt
import networkx as nx
from IPython import display
%matplotlib inline
```

```python
num_individuals = 9
L = util.create_paper_graph(num_individuals)
util.draw_paper_graph(L)
```

```python
def test_leader_locations(loc1, loc2, L):
    x0 = np.zeros([num_individuals, 1])
    leader_indices = [loc1, loc2]
    leader_betas = [1,1]
    beta = util.create_leaderfollower_beta(num_individuals, leader_indices, leader_betas)
    sigma = np.zeros([num_individuals, 1])
    dt = 0.001
    T = 10
    plt.figure(0)
    x = util.run_simulation(x0, beta, L, dt, sigma, T, title=r"Leaders at indices {0} and {1}".format(loc1, loc2))
    plt.figure(1)
    util.plot_discrepancy(x, [loc1, loc2], L, T, ylim=1.5, title=r"Leaders at indices {0} and {1}".format(loc1, loc2), savename="papergraph_leaderlocations_{0}{1}_connected".format(loc1, loc2))
```

```python
test_leader_locations(2,4,L)
```

```python
test_leader_locations(1,2,L)
```
