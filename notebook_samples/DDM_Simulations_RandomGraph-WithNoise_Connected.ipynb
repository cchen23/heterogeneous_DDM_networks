```python
import numpy as np
import DDM_util as util
import networkx as nx
from IPython import display
%matplotlib inline
```

```python
num_individuals = 10
p = 0.5
L = util.create_erdosreyni_graph(num_individuals, p)
util.draw_circle_graph(L, "graphdrawing_random_noisedemo")
```

```python
num_individuals = 10
x0 = np.zeros([num_individuals, 1])
sigma = np.zeros([num_individuals, 1])
sigmaval = 0.03
sigmaval_filename = "03"
sigma_noise = np.ones([num_individuals, 1]) * sigmaval
```

## One Leader

```python
leader_indices = [5]
leader_betas = [1]
beta = util.create_leaderfollower_beta(num_individuals, leader_indices, leader_betas)
dt = 0.001
T = 100
x_oneleader = util.run_simulation(x0, beta, L, dt, sigma, T, title="Opinion States, No Noise", savename="simulation_randomgraph_leader5_nonoise")
```

```python
util.plot_discrepancy(x_oneleader, [5], L, T, title="Discrepancy, No Noise", savename="simulation_randomgraph_leader5_nonoise_discrepancy")
```

```python
leader_indices = [5]
leader_betas = [1]
beta = util.create_leaderfollower_beta(num_individuals, leader_indices, leader_betas)
dt = 0.001
T = 100
x_oneleader_noise = util.run_simulation(x0, beta, L, dt, sigma_noise, T, title=r"Opinion States, $\sigma=0.1$", savename="simulation_randomgraph_leader5_noise")
```

```python
util.plot_discrepancy(x_oneleader_noise, [5], L, T, title=r"Discrepancy, $\sigma=0.1$", savename="simulation_randomgraph_leader5_noise_discrepancy")
```

## Two Leaders

```python
leader_indices = [3,5]
leader_betas = [1,-1]
beta = util.create_leaderfollower_beta(num_individuals, leader_indices, leader_betas)
dt = 0.001
T = 10
x_twoleaders_35_different_nonoise = util.run_simulation(x0, beta, L, dt, sigma, T, ylim=3, title="Opinion States, No Noise", savename="simulation_randomgraph_leader35_diff_nonoise_connected")
```

```python
leader_indices = [3,5]
leader_betas = [1,-1]
beta = util.create_leaderfollower_beta(num_individuals, leader_indices, leader_betas)
dt = 0.001
T = 10
x_twoleaders_35_different_noise = util.run_simulation(x0, beta, L, dt, sigma_noise, T, ylim=3, title=r"Opinion States, $\sigma=%0.2f$" % sigmaval, savename="simulation_randomgraph_leader35_diff_noise%s_connected" % sigmaval_filename)
```
