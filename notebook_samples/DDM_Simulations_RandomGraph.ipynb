## One Leader

```python
num_individuals = 10
x0 = np.zeros([num_individuals, 1])
leader_indices = [5]
leader_betas = [1]
beta = util.create_leaderfollower_beta(num_individuals, leader_indices, leader_betas)
sigma = np.zeros([num_individuals, 1])
dt = 0.001
T = 20
x_oneleader = util.run_simulation(x0, beta, L, dt, sigma, T)
```

```python
util.compute_ydot(L, beta);
```
