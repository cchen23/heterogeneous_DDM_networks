```python
import numpy as np
import DDM_util as util
import matplotlib.pyplot as plt
import networkx as nx
from IPython import display
%matplotlib inline
```

```python
num_individuals = 10
p = 0.5
#L = util.create_erdosreyni_graph(num_individuals, p)
L = util.create_fullyconnected_graph(num_individuals)
graphname = "Fully Connected"
graphfilename = "fc"
```

```python
def test_leader_betas(beta1, beta2):
    x0 = np.zeros([num_individuals, 1])
    leader_indices = [1,2]
    leader_betas = [beta1, beta2]
    beta = util.create_leaderfollower_beta(num_individuals, leader_indices, leader_betas)
    sigma = np.zeros([num_individuals, 1])
    dt = 0.001
    T = 10
    plt.figure(0)
    x = util.run_simulation(x0, beta, L, dt, sigma, T, title=r"{2} graph, leaders at indices 1 and 2, $\beta_1={0}, \beta_2={1}, \sigma=0$".format(graphname, beta1, beta2))
    plt.figure(1)
    util.plot_discrepancy(x, [1,2], L, T, ylim=5, title=r"$\beta_1={0}, \beta_2={1}, \sigma=0$".format(beta1, beta2), savename="twoleader_{0}{1}_{2}_connected".format(beta1, beta2, graphfilename))
```

```python
test_leader_betas(1,1)
```

```python
test_leader_betas(1,2)
```

```python
test_leader_betas(1,5)
```

```python
test_leader_betas(1,-1)
```

```python
test_leader_betas(1,-2)
```
