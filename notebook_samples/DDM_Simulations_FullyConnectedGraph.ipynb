```python
import numpy as np
import DDM_util as util
import networkx as nx
from IPython import display
%matplotlib inline
```

```python
num_individuals = 10
L = util.create_fullyconnected_graph(num_individuals)
util.draw_circle_graph(L)
```

## Homogeneous

```python
x0 = np.zeros([num_individuals, 1])
beta = util.create_homogeneous_beta(num_individuals)
sigma = np.zeros([num_individuals, 1])
dt = 0.001
T = 10
x_homogeneous = util.run_simulation(x0, beta, L, dt, sigma, T)
```

## One Leader

```python
num_individuals = 10
x0 = np.zeros([num_individuals, 1])
leader_indices = [5]
leader_betas = [1]
beta = util.create_leaderfollower_beta(num_individuals, leader_indices, leader_betas)
sigma = np.zeros([num_individuals, 1])
dt = 0.001
T = 20
x_oneleader = util.run_simulation(x0, beta, L, dt, sigma, T)
```
