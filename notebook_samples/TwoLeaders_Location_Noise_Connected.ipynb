```python
import numpy as np
import DDM_util as util
import matplotlib.pyplot as plt
import networkx as nx
from IPython import display
%matplotlib inline
```

```python
num_individuals = 9
L = util.create_paper_graph(num_individuals)
util.draw_paper_graph(L)
```

```python
def test_leader_locations(loc1, loc2, save=True):
    x0 = np.zeros([num_individuals, 1])
    leader_indices = [loc1, loc2]
    sigma_val = 0.01
    sigma_val_filename = str(sigma_val).replace(".","")
    leader_betas = [1,1]
    beta = util.create_leaderfollower_beta(num_individuals, leader_indices, leader_betas)
    sigma = np.ones([num_individuals, 1]) * sigma_val
    dt = 0.001
    T = 20
    plt.figure(0)
    x = util.run_simulation(x0, beta, L, dt, sigma, T, title=r"Leaders at Indices {0} and {1}, $\sigma={2}$".format(loc1, loc2, sigma_val))
    plt.figure(1)
    if save:
        savename="papergraph_leaderlocations_{0}{1}_sigma{2}_discrepancy_connected".format(loc1, loc2, sigma_val_filename)
    else:
        savename=None
    util.plot_discrepancy(x, [loc1, loc2], L, title=r"Leaders at Indices {0} and {1}, $\sigma={2}$".format(loc1, loc2, sigma_val), savename=savename)
```

```python
test_leader_locations(2,4)
```
